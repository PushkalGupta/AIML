```python
df3 = df.groupby('Continent')[['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending = False)
#Can also do [df.columns[5:13]] to save time
df3 = df3.transpose()
df3
```

```python
#But we have taken data from 2022 to 1970 we want growth, so 1970 to 2022
```

```python
df4 = df.groupby('Continent')[df.columns[5:13]].mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
df4 = df4.transpose()
df4 = df4.sort_index(ascending=True)
df4
```
